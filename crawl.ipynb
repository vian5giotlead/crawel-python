```python
import requests
from bs4 import BeautifulSoup

res=requests.get("https://www.cwa.gov.tw/V8/C/W/week.html")
res.encoding = 'utf-8'
print(res.status_code)
if(res.status_code == 200):
    print("OK")
    htmlContent = res.text
    soup = BeautifulSoup(htmlContent, "html.parser")
    rootElement = soup.find('a', attrs={'title':'顯示全臺'})
    print(rootElement)
    print(rootElement['href'])
else:
    print("ERROR")
```

```python
newUrl = "https://stackoverflow.com/questions/55539930/how-to-send-a-soap-request-in-javascript-like-in-soapui"
anRes = requests.get(newUrl)
print(anRes.status_code)
if(anRes.status_code == 200):
    print("OK")
    anContent = anRes.text
    anSoup = BeautifulSoup(anContent, "html.parser")
    print(anSoup)
    print(anSoup.title.text)
else:
    print("ERROR")
```
